```python
import pandas as pd
from datetime import date

```

```python
df = pd.read_csv("https://cli.fusio.net/cli/climate_data/webdata/mly532.csv", skiprows=19, low_memory=False) #nrows=100
#df_index = pd.read_csv("https://cli.fusio.net/cli/climate_data/webdata/mly532.csv", nrows=20)
#df_index
df
```

```python
df['date_time'] = pd.to_datetime(df[['year', 'month']].assign(DAY=1)) #Combine month year columns to a new timedate column. https://stackoverflow.com/questions/48304927/cleanly-combine-year-and-month-columns-to-single-date-column-with-pandas
df
```

```python
df['date_ordinal'] = pd.to_datetime(df['date_time']).apply(lambda date: date.toordinal())
df
```

```python
#seaborn lineplot
import seaborn as sns
sns.lineplot(data=df[0:], x="date_time", y="rain")
#sns.regplot(data=df[0:], x="date_ordinal", y="rain")
```

```python
#seaborn lineplot regression line. Date_ordinal added to df. X axis label changed
import seaborn as sns
title = 'Dublin Airport monthly mean temperature 1941-2022'
ax = sns.set(rc={"figure.figsize":(12, 10)}) #width=6, height=5 #set size of plot
ax = sns.lineplot(data=df[0:], x="date_ordinal", y="meant").set(title= title )
ax = sns.regplot(data=df[0:], x="date_ordinal", y="meant")
ax.set_xlabel('date_time')
new_labels = [date.fromordinal(int(item)) for item in ax.get_xticks()]
ax.set_xticklabels(new_labels)
```
